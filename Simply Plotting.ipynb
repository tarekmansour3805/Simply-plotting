```python
print('Line plots show data over time')
plt.plot(neighbourhood.iloc[0:10000]['number_of_reviews'])
plt.show()
```

```python
print('reviews and price breakdowns:')
plt.scatter(neighbourhood['number_of_reviews'],neighbourhood['price'])
plt.show()
```

```python
print('Hotel Reviews:')
plt.hist(neighbourhood['number_of_reviews'])
plt.show()
```

```python
print('Price chart:')
norm= np.random.normal(0, 1, 140)
norm.sort()
plt.plot(norm, neighbourhood['price'].sort_values(), "o") 
plt.show() 
```
