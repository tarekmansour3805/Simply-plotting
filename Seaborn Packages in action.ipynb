```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
best= pd.read_csv('listings 2.csv')
```

```python
plt.figure(figsize=(20,9))
best.groupby(['price'])['number_of_reviews'].sum().plot()
plt.ylabel('Reviews')
```

```python
#plt.scatter(figsize=(20,9))
#df.groupby(['price'])['number_of_reviews'].sum().plot()
```

```python
x = best['price']
y = best['number_of_reviews']
plt.title('Price and reviews chart:')
plt.xlabel('Price')
plt.ylabel('Reviews')

plt.scatter(x, y);



plt.show()
```

```python
sns.FacetGrid(x="price", y="number_of_reviews", hue="Best",
            kind="line", data=best);
```

```python
sns.relplot(x="price", y="number_of_reviews", 
            col="align", hue="choice", 
            size="coherence", style="choice", 
            facet_kws=dict(sharex=False), kind="line", 
            legend="full", data=best);
```
